### Python as a calculator
python을 활용해 사칙연산 뿐만 아니라 더 심화된 계산도 할 수 있음
- Exponentiation: ** (ex. 4 ** 2 실행 결과는 16)
- Modulo: % (나눗셈의 나머지 계산, ex. 18 % 7 실행 결과는 4)

```python
# Addition, subtraction
print(5 + 5)
print(5 - 5)

# Multiplication, division, modulo, and exponentiation
print(3 * 5)
print(10 / 2)
print(18 % 7)
print(4 ** 2)
```

### Variable assignment
변수를 지정하고 나면, 이후에 변수 이름을 활용해서 그 값을 불러올 수 있다.  
변수는 "=" 기호를 활용해 지정 (Python에서 "="는 assignment의 의미. equality를 의미하는 것이 아니다!)

```python
# Create a variable savings
savings = 100

# Print out savings
print(savings)
```

### Calculations with variables

```python
# $100를 갖고 있고, 1년마다 10%의 이자를 받음. 7년 후 갖게 될 금액은?
# Create a variable savings
savings = 100

# Create a variable growth_multiplier
growth_multiplier = 1.1

# Calculate result
result = savings * growth_multiplier ** 7

# Print out result
print(result)
```

### Operations with Other Type
Different types behave differently in Python!  

```python
# +가 int변수와 str변수에서 서로 다르게 역할
savings = 100
desc = "compound interest"

savings2 = savings + savings
print(savings2)
desc2 = desc + desc
print(desc2)
```

### Type Conversion
서로 다른 타입의 변수들로 그냥 연산을 할 수는 없다.  

Type converting functions
- int() : int로 바꿔줌
- float() : float로 바꿔줌
- str() : str으로 바꿔줌
- bool() : bool으로 바꿔줌

```python
# Definition of savings and result
savings = 100
result = 100 * 1.10 ** 7
print(type(savings))
print(type(result))

# type 바꿔주지 않고 연산하면 오류 발생
print("I started with $" + savings + " and now have $" + result + ". Awesome!")
```

```python
# str과 연산이 가능하도록 savings와 result의 type을 str로 바꿔준다
print("I started with $" + str(savings) + " and now have $" + str(result) + ". Awesome!")
```
