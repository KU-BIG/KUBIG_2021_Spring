```python
import pandas as pd
data = pd.read_csv("titanic.csv")
```

```python
import numpy as np
titanic = np.array(data)
titanic
```

titanic의 type과 shape을 각각 출력하여라

```python
print(type(titanic))
print(titanic.shape)
```

25번째 승객의 이름을 출력하여라

```python
titanic[24, 3]
```

승객들의 이름만 저장한 name 변수를 만들어라(이름은 4번째 열)

```python
name = titanic[:,3]
print(name)
```

승객들의 나이만 저장한 age 변수를 만들어라(나이는 6번째 열)

```python
age = titanic[:,5]
print(age)
```

승객들의 나이의 중간값을 구하여라

```python
np.median(age)
```
