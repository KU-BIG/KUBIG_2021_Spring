## 1) NumPy array

#### numpy 배열 생성의 대표적인 방법
#### np.array() 함수 안에 list를 넣어 배열을 생성한다.

```python
# example.array

baseball = [180, 215, 210, 210, 188, 176, 209, 200]   # baseball 이라는 list 생성
import numpy as np                                   # numpy를 호출해 np로 지정 
np_baseball=np.array(baseball)                        # np.array 함수를 사용해 배열 생성 
print(np_baseball)
```

```python
# example.2Darray
baseball = [[180, 78.4],
            [215, 102.7],
            [210, 98.5],
            [188, 75.2]]
import numpy as np
np_baseball=np.array(baseball)
print(np_baseball.shape)         #shape: 배열의 속성 보여줌
print(np_baseball.ndim)          #ndim: 차원을 보여줌
print(np_baseball.size)          #size: 원소의 총 개수를 보여줌
print(np_baseball) 
```

## 2) NumPy에서의 연산

#### numpy를 사용하는 이유
#### list 형태에서는 파이썬이 일반적인 계산을 실행할 방법을 모르기 때문에 error가 뜸 (예제 스킵)
#### 반면 numpy를 사용할 경우 리스트 안의 각 원소에 대한 연산으로 인식

```python
# example.arithmetic
import numpy as np
np_mat = np.array([[1, 2],
                   [3, 4],
                   [5, 6]])
mat=([[1, 2],
      [3, 4],
      [5, 6]])
print(mat*2)
print(np_mat * 2)
print(np_mat + np.array([10, 10]))
print(np_mat + np_mat)
```

## 3) Boolean NumPy array
#### 배열 각 요소의 선택여부를 True, False로 표현하는 방식  
#### 특정 요소만을 추출하고 싶을 때 사용 가능

```python
# example.boolean
import numpy as np
lst = [                        # 3X3의 list 생성:1st
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
a = np.array(lst)
bool_indexing = (a % 2 == 0)   # 배열 a 에 대해 짝수면 True, 홀수면 False (%: 나머지 계산)
print(bool_indexing)
print(a[bool_indexing])        # boolean 인덱스를 사용하여 True인 요소만 뽑아냄

n = a[ a % 2 == 0 ]            # 더 간단한 표현
print(n)
```

## 4) NumPy 슬라이싱
#### numpy 배열은 파이썬 리스트와 마찬가지로 슬라이스(Slice)를 지원
#### 각 차원별로 슬라이스 범위를 지정

```python
# example.slice
import numpy as np
lst = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
arr = np.array(lst)
a = arr[0:2, 0:2]   # 행(0,1)과 열(0,1)를 호출
print(a)
a = arr[1:, 1:]     # 행(2,3)과 열(2,3)을 호출
print(a)
```

## 3) NumPy 분할/덧붙이기
#### ㄱ) 분할: 한 개의 자료를 두 개 이상으로 나누는 과정
#### - 머신러닝에서, 자료를 학습데이터(training data)와 시험데이터(test data)로 나누어 모형을 학습데이터에 적합시키고 이 적합된 모형을    시험데이터에 적용 (실제로는 sklearn model_selection의 train_test_split을 많이 사용)
#### ex) np.split(), np.vsplit(), np.hsplit()
#### ㄴ) 덧붙이기: 두 개 이상의 자료를 하나의 자료로 합치는 과정
#### - 자료의 분석을 위해 자료를 합쳐야 하는 경우 발생
#### ex) np.concatenate(), np.hstack(), np.vstack()

```python
# exercise.분할

x=[1,2,3,4,5,6,7,8]
x1,x2,x3=np.split(x,[2,6])   # index를 [0,2), [2,6), [6,마지막 index]로 분류
print(x1,x2,x3)

x=np.random.random((4,4))   
x1,x2=np.vsplit(x,[2])      #배열을 행 방향으로 분류
print(x)
print(x1)
print(x2)
```

```python
# exercise.덧붙이기

x=np.array([1,2])
y=np.array([2,3])
z=np.array([4,5])
xyz=np.concatenate([x,y,z])
print(xyz)

xx=np.array([[1,2,3],[4,5,6]])
yy=np.array([[2,4,6],[8,7,6]])
xxyy=np.concatenate([xx,yy])          # np.vstack()
print(xxyy)

xxyy=np.concatenate([xx,yy],axis=1)   # np.hstack()
print(xxyy)
```
