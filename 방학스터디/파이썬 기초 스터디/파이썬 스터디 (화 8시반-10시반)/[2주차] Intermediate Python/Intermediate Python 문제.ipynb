##### 예시) 주사위 던지기 게임

- 주사위를 총 100번 던짐
- 1이나 2가 나오면 한 계단 내려감
- 3,4,5가 나오면 한 계단 올라감
- 6이 나오면 주사위를 한 번 더 던져 나오는 결과만큼 올라감

<주의사항>
- 계단 0 밑으로는 내려갈 수 없음
- 계단에서 떨어지면 0에서부터 다시 시작 (떨어질 확률: 0.1%)

<분포>
- 500번의 시뮬레이션 결과 60계단 이상 올라갈 확률을 히스토그램으로 나타냄

```python
###예시###

all_walks = []                         # 시뮬레이션을 위한 list 생성

for i in range(500) :                  # 게임을 500번 시행
    random_walk = [0]                  # 주사위의 결과를 넣을 list의 초기값:0
    
    for x in range(100) :              # 100번 주사위를 던짐
        step = random_walk[-1]         # append를 사용할 예정
        dice = np.random.randint(1,7)  # 주사위 난수 생성 
        
        ## if 문 ##
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        if np.random.rand() < 0.001 :   # 임의로 생성한 실수가 0.001보다 작다면 step=0으로 초기화
            step = 0

        random_walk.append(step)        # append를 사용해 주사위의 결과를 list에 넣어줌
    
    all_walks.append(random_walk)       # append를 사용해 게임의 결과를 시뮬레이션 list에 넣어줌

np_aw_t = np.transpose(np.array(all_walks))  # row: x축. observation: y축

ends = np_aw_t[-1,:]                    # np_aw_t의 마지막 행(최종 step 값)을 ends로 지정

plt.hist(ends)                          # ends의 히스토그램을 그려줌
plt.show()

print(np.mean(ends > 60))               # 60보다 클 경험적 확률 출력
```
