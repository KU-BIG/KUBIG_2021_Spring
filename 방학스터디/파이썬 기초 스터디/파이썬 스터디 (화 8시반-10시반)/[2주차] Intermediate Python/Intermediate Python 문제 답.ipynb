```python
'Asia' = red \
'Europe' =  green \
'Africa' = blue \
'Americas' = yellow \
'Oceania' = black
```

```python
n = gapminder2007.shape[0]
color = []
for i in range(n):
  if gapminder2007['continent'][i] == 'Asia' :
    color.append('red')
  elif gapminder2007['continent'][i] == 'Europe' :
    color.append('green')
  elif gapminder2007['continent'][i] == 'Africa' :
    color.append('blue')
  elif gapminder2007['continent'][i] == 'Americas' :
    color.append('yellow')
  else :
    color.append('black')
```

```python
import matplotlib.pyplot as plt

# Specify c and alpha inside plt.scatter()
plt.scatter(x = gdp_cap, y = life_exp, s = np_pop, c = color, alpha = 0.8)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000], ['1k','10k','100k'])

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123) 




all_balance = []

for i in range(1000) :                   # 게임을 1000번 시행(simulation)
    balance = [3]
    
    for x in range(100) :                # 100번의 동전던지기 게임
        result=balance[-1]          
        coin = np.random.randint(0,2)    
        
        if coin == 0 :
            result = result - 1          # 뒷면이 나오면 -1                    
        else :
            result = result + 2*np.random.randint(0,2)   # 앞면이 나오면 동전을 한 번 더 던져서 나온 결과에 x2

        if np.random.rand() < 0.001 :   # 임의로 생성한 실수가 0.001보다 작다면 result=3으로 초기화
            result = 3

        balance.append(result)
    
    all_balance.append(balance)

np_ab=np.array(all_balance) 
np_ab_t=np.transpose(np_ab)

ends = np_ab_t[-1,:]                    # np_ab_t의 마지막 행(최종 result 값)을 ends로 지정

plt.hist(ends)                          # ends의 히스토그램을 그려줌
plt.show()




decision=np.mean(ends > 3)                # 도박꾼의 balance가 3보다 클 확률
if decision>=0.5:
    print('도전!')
else: 
    print('포기ㅠ')
    
print(decision)
```
