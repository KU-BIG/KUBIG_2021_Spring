# Multiple Parameters and Return Values

```python
#A Multiple Paramter Function

def raise_to_power(value1, value2):
    """Raise value1 to the power of value2."""
    new_value = value1 ** value2
    return new_value
```

```python
result= raise_to_power(3,4)

print(result)
```

### A Multiple Return-Values Function: Using 'Tuple'

#### Tuples?
 - Like a list: Can contain multiple values.
 
But...
 - Immutable: Cannot modify values.  (not like a 'list')
 - Constructed using parantheses ().  (note that 'list' uses [])

```python
even_nums = (2,4,6)
print(type(even_nums))
```

```python
#Multiple Parameter, Multiple Variable Returns

def raise_both(value1, value2):
    """Raise value1 to the power of value2 and vice versa."""
    new_value1 = value1 ** value2
    new_value2 = value2 ** value1
    
    new_tuple = (new_value1, new_value2)
    
    return new_tuple
```

```python
raise_both(3,4)
```

# Basic Ingredients of a Function

### Function Header

def raise_both(value1, value2):  <- Function Header
  => def + function name + (parameters) + :

### Function Body
#### """Raise value1 to the power of value2 and vice versa.""" <- docstrings
####   new_value1 = value1 ** value2        <- computation starts
####   new_value2 = value2 ** value1
    
####   new_tuple = (new_value1, new_value2)    <- computation ends
    
####   return new_tuple <- closes with return or print

### Recap:
To add 1 to the value of a key entry in a dictionary d, use d[entry] += 1. To add a new key to d and set its value to 1, use d[entry] = 1.

```python
# Import pandas
import pandas as pd

# Import Twitter data as DataFrame: df
df = pd.read_csv('tweets.csv')

# Initialize an empty dictionary: langs_count
langs_count = {}

# Extract column from DataFrame: col
col = df['lang']

# Iterate over lang column in DataFrame
for entry in col:

    # If the language is in langs_count, add 1
    if entry in langs_count.keys():
        langs_count[entry] += 1
    # Else add the language to langs_count, set the value to 1
    else:
        langs_count[entry] = 1

# Print the populated dictionary
print(langs_count)
```
