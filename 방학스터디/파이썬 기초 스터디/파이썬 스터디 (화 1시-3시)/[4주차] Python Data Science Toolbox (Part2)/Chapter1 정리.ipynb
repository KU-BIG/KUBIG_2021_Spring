**Iterator**

**Iterable:** An object with an associated iter() method

즉, iter()를 interable에 사용하면 'iterator'가 됨

list, dictionaries, tuple, string, file connections 등 적용가능

**Iterator:** Produces next value with next()

```
#iterable: iter 함수를 사용하여 iterator 객체를 반환 할 수 있는 객체
#Iterator: next 함수를 사용하여 다음 객체를 반복적으로 하나씩 가져오는 동작을 할 수 있는 객체
```

```
a = [1,2,3]
print(a.__next__) #그냥 리스트에 적용시키면 에러
```

```
a = iter(a)
print(a.__next__())
```

```
for letter in 'Datacamp': #for를 사용해서도 반복가능 #for문을 사용하면 iterable 객체를 임시적으로 iterator로 바꿀 수 있음(?)
  print(letter)
```

```
for i in range(4): #이런식 
  print(i)
```

```
# Using Enumerate
```

```
mutants = ['charles xavier','bobby drake','kurt wagner','max eisenhardt','kitty pryde']

mutant_list = list(enumerate(mutants)) #enumerate를 list로 만들기  

print(mutant_list)
```

```
for index1, value1 in enumerate(mutants):
  print(index1, value1)
```

```
for index2, value2 in enumerate(mutants, start=10): #start 기능을 통해서 index 숫자 수정 
  print(index2, value2)
```

```
# Using Zip
```

```
mutants = ['charles xavier','bobby drake','kurt wagner','max eisenhardt','kitty pryde']
aliases = ['prof x', 'iceman', 'nightcrawler', 'magneto', 'shadowcat']
powers = ['telepathy','thermokinesis','teleportation','magnetokinesis','intangibility']
```

```
mutant_zip = zip(mutants, aliases, powers)

print(mutant_zip) #zip만 사용하면 의미 없음 
```

```
mutant_data = list(zip(mutants, aliases, powers))

# Print the list of tuples

print(mutant_data)
```

```
for value1, value2, value3 in mutant_zip:
    print(value1, value2, value3)
```

```
# Using * and zip to 'unzip'
```

```
# Create a zip object from mutants and powers: z1
z1 = zip(mutants, powers)

print(z1)  
```

```
# Print the tuples in z1 by unpacking with *
print(*z1)    #직접적으로 unzip하는 기능은 없음 * 사용 
```

```
z1 = zip(mutants, powers)

# 'Unzip' the tuples in z1 by unpacking with * and zip(): result1, result2 #다른방식으로 
result1, result2 = zip(*z1)
```

```
# Check if unpacked tuples are equivalent to original tuples
print(result1 == mutants)
print(result2 == powers) #unzip한 것은 튜플로 나옴 (위에는 리스트)
```

```
type(result2) #튜플 확인
```

```
result1 = list(result1)
result2 = list(result2)
print(result1 == mutants)
print(result2 == powers)
```
