```python
#customization 
import numpy as np

# Basic scatter plot, log scale
plt.scatter(gdp_cap, life_exp)
plt.xscale('log') #로그화 !! 

# Strings
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007' #설명 삽입 

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab) #축 설명 

# Add title
plt.title(title) #제목 추가

# After customizing, display the plot
plt.show()
```

```python
# Definition of tick_val and tick_lab

tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k'] #리스트의 길이가 같아야함 

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)

# Specify c and alpha inside plt.scatter()
plt.scatter(x = gdp_cap, y = life_exp, s = np.array(pop) * 2, c = "blue", alpha = 0.5)
#S = SIZE , C = COLOR, ALPHA = 점들의 투명도 

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000], 
           ['1k','10k','100k']) #척도 만들어 주기 

# Additional customizations
plt.text(1550, 71, 'India')
plt.text(5700, 80, 'China')  # 표안에 설명 추가 

# Add grid() call
plt.grid(True)        #GRID 만들어주기 

# Show the plot
plt.show()
```
